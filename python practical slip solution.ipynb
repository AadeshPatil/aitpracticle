```python
#Q.5
def find_pair_indices(numbers, target):
    num_dict = {}
    for i, num in enumerate(numbers):
        complement = target - num
        if complement in num_dict:
            return [num_dict[complement], i]
        num_dict[num] = i
    return []

# Example array of numbers and target
numbers = [2, 7, 11, 15]
target = 9

# Find the pair of elements
pair_indices = find_pair_indices(numbers, target)

# Print the pair indices
if pair_indices:
    print("Pair indices:", pair_indices[0], pair_indices[1])
else:
    print("No pair found with the given target.")
```

```python
#Q.8
class Investment:
    def __init__(self, principal, interest):
        self.principal = principal
        self.interest = interest

    def value_after(self, years):
        value = self.principal * (1 + self.interest) ** years
        return value


# Create an instance of the Investment class
investment = Investment(1000, 0.05)

# Calculate the value after 3 years
value_after_3_years = investment.value_after(3)
print("Value after 3 years:", value_after_3_years)

# Calculate the value after 5 years
value_after_5_years = investment.value_after(5)
print("Value after 5 years:", value_after_5_years)
```

```python
#Q12. Write a Python program to check the validity of a password (input from users).
#Validation :

import re

def validate_password(password):
    if len(password) < 6 or len(password) > 16:
        return False

    if not re.search(r"[a-z]", password):
        return False

    if not re.search(r"[A-Z]", password):
        return False

    if not re.search(r"\d", password):
        return False

    if not re.search(r"[$#@]", password):
        return False

    return True


# Prompt the user to enter a password
password = input("Enter a password: ")

# Validate the password
valid = validate_password(password)

# Display the result
if valid:
    print("Password is valid.")
else:
    print("Password is not valid.")
```

```python
#Q16. Write a programm for synchronization of Threads using RLOCK. Accept the two
#numbers from user and calculate factorial of both Numbers simultaneously
import threading

def calculate_factorial(number, lock):
    result = 1
    for i in range(1, number + 1):
        lock.acquire()
        result *= i
        lock.release()
    print(f"Factorial of {number} is {result}")

# Accept two numbers from the user
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))

# Create an RLock object
lock = threading.RLock()

# Create two threads for calculating factorial of each number
thread1 = threading.Thread(target=calculate_factorial, args=(number1, lock))
thread2 = threading.Thread(target=calculate_factorial, args=(number2, lock))

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

```

```python
#Q17. Write a multithreading program where one thread prints square of a number and
#another thread prints factorial of a number. Also display the total time taken for the
#execution

import threading
import time
import math

def calculate_square(number):
    square = number ** 2
    print(f"Square of {number}: {square}")

def calculate_factorial(number):
    factorial = math.factorial(number)
    print(f"Factorial of {number}: {factorial}")

# Accept the number from the user
number = int(input("Enter a number: "))

# Create a timer to measure execution time
start_time = time.time()

# Create two threads for calculating square and factorial
thread1 = threading.Thread(target=calculate_square, args=(number,))
thread2 = threading.Thread(target=calculate_factorial, args=(number,))

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

# Calculate the total execution time
total_time = time.time() - start_time
print(f"Total time taken: {total_time} seconds")
```

```python
#Q18. Write a python program to check the given number is prime or not. Handle Suitable
#Exception
def is_prime(number):
    if number < 2:
        return False

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False

    return True

try:
    number = int(input("Enter a number: "))
    if is_prime(number):
        print(f"{number} is a prime number.")
    else:
        print(f"{number} is not a prime number.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
except Exception as e:
    print("An error occurred:", e)
```

```python
#Q19. Develop a program to Print the number of line,words,and characters present in the
#given file. Accept the file name from user. Handle necessary exceptions
def count_file_stats(file_name):
    try:
        with open(file_name, 'r') as file:
            lines = 0
            words = 0
            characters = 0

            for line in file:
                lines += 1
                words += len(line.split())
                characters += len(line)

            print(f"Number of lines: {lines}")
            print(f"Number of words: {words}")
            print(f"Number of characters: {characters}")

    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
    except PermissionError:
        print(f"Permission denied to open file '{file_name}'.")
    except Exception as e:
        print("An error occurred:", e)


# Accept the file name from the user
file_name = input("Enter the file name: ")

# Call the function to count file statistics
count_file_stats(file_name)
```

```python
#Q20. Python Program to Append the Contents of One File to Another File
def append_file(source_file, destination_file):
    try:
        with open(source_file, 'r') as source, open(destination_file, 'a') as destination:
            contents = source.read()
            destination.write(contents)

        print("File contents appended successfully.")
    except FileNotFoundError:
        print("One or both files not found.")
    except PermissionError:
        print("Permission denied to open file(s).")
    except Exception as e:
        print("An error occurred:", e)


# Accept the file names from the user
source_file = input("Enter the source file name: ")
destination_file = input("Enter the destination file name: ")

# Call the function to append file contents
append_file(source_file, destination_file)
```

```python
#Q21. Develop a program to remove the comment character from all the lines in a given file.
#Accept file name from user
def remove_comments(file_name):
    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()

        # Remove comment character from each line
        lines = [line.split('#', 1)[0] for line in lines]

        with open(file_name, 'w') as file:
            file.writelines(lines)

        print("Comments removed successfully.")
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
    except PermissionError:
        print(f"Permission denied to open file '{file_name}'.")
    except Exception as e:
        print("An error occurred:", e)


# Accept the file name from the user
file_name = input("Enter the file name: ")

# Call the function to remove comments from the file
remove_comments(file_name)
```

```python
#Q23. Write a Python program to combine each line from first file with the corresponding
#line in second file.
def combine_files(file1, file2, output_file):
    try:
        with open(file1, 'r') as file1, open(file2, 'r') as file2, open(output_file, 'w') as output:
            lines1 = file1.readlines()
            lines2 = file2.readlines()

            # Combine lines from both files and write to the output file
            for line1, line2 in zip(lines1, lines2):
                combined_line = line1.strip() + " " + line2.strip() + "\n"
                output.write(combined_line)

        print("Files combined successfully.")
    except FileNotFoundError:
        print("One or more files not found.")
    except PermissionError:
        print("Permission denied to open file(s).")
    except Exception as e:
        print("An error occurred:", e)


# Provide the file names
file1 = "file1.txt"
file2 = "file2.txt"
output_file = "combined.txt"

# Call the function to combine the files
combine_files(file1, file2, output_file)

```

```python
"""Q32. Write a Python program to convert two date difference in days, hours, minutes,
seconds"""
from datetime import datetime

start_date_str = input("Enter the start date (YYYY-MM-DD HH:MM:SS): ")
end_date_str = input("Enter the end date (YYYY-MM-DD HH:MM:SS): ")

start_date = datetime.strptime(start_date_str, "%Y-%m-%d %H:%M:%S")
end_date = datetime.strptime(end_date_str, "%Y-%m-%d %H:%M:%S")

time_diff = end_date - start_date

days = time_diff.days
hours, remainder = divmod(time_diff.seconds, 3600)
minutes, seconds = divmod(remainder, 60)

print("Difference: {} days, {} hours, {} minutes, {} seconds".format(days, hours, minutes, seconds))
```

```python
"""Q34. Write a python code for handling missing data using fillna(),dropna() method on
Iris.csv file[fill the data using mean/median/mode value]
"""
import pandas as pd

# Load the Iris dataset
data = pd.read_csv('iris.csv')

# Display the number of missing values in each column
print("Number of missing values before handling:")
print(data.isnull().sum())

# Fill missing values with mean
data_filled_mean = data.fillna(data.mean())

# Fill missing values with median
data_filled_median = data.fillna(data.median())

# Fill missing values with mode
data_filled_mode = data.fillna(data.mode().iloc[0])

# Drop rows with missing values
data_dropped = data.dropna()

# Display the number of missing values in each column after handling
print("\nNumber of missing values after handling:")
print(data_filled_mean.isnull().sum())

# Display the filled datasets
print("\nData with missing values filled using mean:")
print(data_filled_mean.head())

print("\nData with missing values filled using median:")
print(data_filled_median.head())

print("\nData with missing values filled using mode:")
print(data_filled_mode.head())

# Display the dropped dataset
print("\nData with missing values dropped:")
print(data_dropped.head())
```

```python
"""Q36. Use Bollywood.csv file to answer this questions:
a) How many records are present in the dataset? Print the metadata information of the
dataset.
b)How many movies got released in each genre? Which genre had highest number of
releases? Sort number of releases in each genre in descending order.
c) Which month of the year, maximum, number movie releases are seen?(Note:
Extract a new column called month from releaseDate column)
"""

import pandas as pd

# a) How many records are present in the dataset? Print the metadata information of the dataset.
df = pd.read_csv('Bollywood.csv')
num_records = len(df)
print("Number of records:", num_records)
print("\nMetadata information:")
print(df.info())

# b) How many movies got released in each genre? Which genre had the highest number of releases? Sort number of releases in each genre in descending order.
genre_counts = df['Genre'].value_counts()
print("\nNumber of releases in each genre:")
print(genre_counts)
max_genre = genre_counts.idxmax()
print("\nGenre with the highest number of releases:", max_genre)

# c) Which month of the year had the maximum number of movie releases? (Note: Extract a new column called 'month' from the 'releaseDate' column)
df['month'] = pd.to_datetime(df['releaseDate']).dt.month
month_counts = df['month'].value_counts()
max_month = month_counts.idxmax()
print("\nMonth with the maximum number of movie releases:", max_month)
```

```python
"""Q38. Use company_sales_data.csv file to answer this questions:
a. Read Total profit of all months and show it using a line plot
b. Get total profit of all months and show line plot with the following Style properties:
i. Line Style dotted and Line-color should be red
ii. Show legend at the lower right location.
iii. X label name = Month Number
iv. Y label name = Sold units number
"""

import pandas as pd
import matplotlib.pyplot as plt

# a) Read Total profit of all months and show it using a line plot
df = pd.read_csv('company_sales_data.csv')
months = df['month_number']
profits = df['total_profit']
plt.plot(months, profits)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.show()

# b) Get total profit of all months and show line plot with the specified style properties
plt.plot(months, profits, linestyle='dotted', color='red')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')
plt.legend(['Total Profit'], loc='lower right')
plt.show()
```

```python
"""Q39. Use company_sales_data.csv file to answer this questions:
a. Read all product sales data and show it using a multiline plot
b. Read face cream and facewash product sales data and show it using the bar chart
c. Calculate total sale data for last year for each product and show it using a Pie chart
"""
import pandas as pd
import matplotlib.pyplot as plt

# a) Read all product sales data and show it using a multiline plot
df = pd.read_csv('company_sales_data.csv')
months = df['month_number']
face_cream_sales = df['facecream']
face_wash_sales = df['facewash']
toothpaste_sales = df['toothpaste']
bathing_soap_sales = df['bathingsoap']
shampoo_sales = df['shampoo']
moisturizer_sales = df['moisturizer']

plt.plot(months, face_cream_sales, label='Face Cream')
plt.plot(months, face_wash_sales, label='Face Wash')
plt.plot(months, toothpaste_sales, label='Toothpaste')
plt.plot(months, bathing_soap_sales, label='Bathing Soap')
plt.plot(months, shampoo_sales, label='Shampoo')
plt.plot(months, moisturizer_sales, label='Moisturizer')

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Product Sales Data')
plt.legend()
plt.show()

# b) Read face cream and facewash product sales data and show it using a bar chart
plt.bar(months, face_cream_sales, label='Face Cream')
plt.bar(months, face_wash_sales, label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.title('Face Cream and Face Wash Sales Data')
plt.legend()
plt.show()

# c) Calculate total sale data for last year for each product and show it using a Pie chart
last_year_sales = df.loc[df['year'] == df['year'].max()]
last_year_sales = last_year_sales.drop(columns=['month_number', 'year', 'total_units', 'total_profit'])
product_totals = last_year_sales.sum()

product_names = product_totals.index
product_sales = product_totals.values

plt.pie(product_sales, labels=product_names, autopct='%1.1f%%')
plt.title('Total Sales for Last Year')
plt.show()
```
